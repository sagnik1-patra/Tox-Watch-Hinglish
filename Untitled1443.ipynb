```python
import os, json, re
from datetime import datetime
from typing import Any, Dict, List, Tuple

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import StratifiedKFold, learning_curve
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import (
    accuracy_score, precision_recall_fscore_support, classification_report,
    confusion_matrix
)

# ------------------ USER PATHS ------------------
TRAIN_PATH = r"C:\Users\sagni\Downloads\Tox Watch Hinglish\archive\Hatespeech-Hindi_Train.csv"
VALID_PATH = r"C:\Users\sagni\Downloads\Tox Watch Hinglish\archive\Hatespeech-Hindi_Valid.csv"
TEST_PATH  = r"C:\Users\sagni\Downloads\Tox Watch Hinglish\archive\Hatespeech-Hindi_Test.csv"
OUT_DIR    = r"C:\Users\sagni\Downloads\Tox Watch Hinglish"
# ------------------------------------------------

RANDOM_STATE = 42

TEXT_CANDS = [
    "text","tweet","sentence","content","message","post","comment",
    "clean_text","utterance","selftext","title"
]
LABEL_CANDS = [
    "label","category","class","target","task_1","task_2","subtask_a","hs_label","y"
]

# --------------- I/O & DETECTION HELPERS ---------------
def ensure_out():
    os.makedirs(OUT_DIR, exist_ok=True)

def read_csv(path: str) -> pd.DataFrame:
    if not os.path.exists(path):
        raise FileNotFoundError(path)
    try:
        return pd.read_csv(path)
    except UnicodeDecodeError:
        return pd.read_csv(path, encoding="latin-1")

def detect_text_and_label(df: pd.DataFrame) -> Tuple[pd.Series, pd.Series, str, str]:
    lower = {c.lower(): c for c in df.columns}

    # text column (or title+selftext fallback)
    tcol = None
    for c in TEXT_CANDS:
        if c in lower:
            tcol = lower[c]; break
    if tcol is None and "title" in lower and "selftext" in lower:
        text = (df[lower["title"]].fillna("").astype(str) + " " +
                df[lower["selftext"]].fillna("").astype(str)).str.strip()
        tcol_used = "title+selftext"
    else:
        if tcol is None:
            obj = [c for c in df.columns if df[c].dtype == object]
            if not obj:
                raise ValueError("No obvious text column. Rename to 'text'.")
            tcol = obj[0]
        text = df[tcol].astype(str)
        tcol_used = tcol

    text = text.fillna("").str.replace(r"\s+"," ", regex=True).str.strip()
    mask = text != ""
    text = text[mask]
    df = df.loc[text.index]

    # label column (low-cardinality fallback)
    lcol = None
    for c in LABEL_CANDS:
        if c in lower:
            lcol = lower[c]; break
    if lcol is None:
        for c in df.columns:
            uniq = pd.Series(df[c].dropna().unique())
            if uniq.size <= 50 and (df[c].dtype == object or pd.api.types.is_integer_dtype(df[c])):
                lcol = c; break
    if lcol is None:
        raise ValueError("No label column found. Rename to 'label'.")

    labels_raw = df[lcol].loc[text.index]
    return text, labels_raw, tcol_used, lcol

def build_label_map(splits_raw: Dict[str, pd.Series]) -> Dict[Any, int]:
    vals = []
    for s in splits_raw.values():
        vals.extend(list(pd.Series(s).dropna().unique()))
    # preserve first-seen order, then sort by string for stability
    vals = list(dict.fromkeys(vals))
    vals_sorted = sorted(vals, key=lambda x: str(x).lower())
    return {v: i for i, v in enumerate(vals_sorted)}

def apply_label_map(series: pd.Series, label2id: Dict[Any, int]) -> pd.Series:
    mapped = series.map(lambda x: label2id.get(x, None))
    return mapped[mapped.notna()].astype(int)

# ------------------ MODEL ------------------
def build_pipeline() -> Pipeline:
    return Pipeline(steps=[
        ("tfidf", TfidfVectorizer(
            lowercase=True,
            ngram_range=(1,2),           # uni+bi
            min_df=2,
            max_features=300_000,
            strip_accents="unicode"
        )),
        ("clf", LogisticRegression(
            solver="saga",
            penalty="l2",
            C=1.0,
            max_iter=300,
            n_jobs=-1,
            random_state=RANDOM_STATE,
            class_weight="balanced",
            multi_class="auto"
        ))
    ])

# ------------------ PLOTS ------------------
def plot_learning_curve(pipe: Pipeline, X, y, out_path: str):
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
    train_sizes, train_scores, val_scores = learning_curve(
        estimator=pipe,
        X=X, y=y,
        cv=cv,
        scoring="accuracy",
        train_sizes=np.linspace(0.1, 1.0, 8),
        n_jobs=-1,
        shuffle=True,
        random_state=RANDOM_STATE
    )
    tr = train_scores.mean(axis=1)
    va = val_scores.mean(axis=1)

    plt.figure(figsize=(7,4))
    plt.plot(train_sizes, tr, marker="o", label="Training accuracy")
    plt.plot(train_sizes, va, marker="s", label="Validation accuracy")
    plt.title("Learning Curve (Accuracy)")
    plt.xlabel("Training samples"); plt.ylabel("Accuracy")
    plt.grid(True, linestyle="--", alpha=0.4)
    plt.legend()
    plt.tight_layout()
    plt.savefig(out_path, dpi=160)
    plt.show()

def plot_confusion_matrix(y_true, y_pred, id2label: Dict[int,str], out_path: str):
    labels = sorted(list(set(y_true) | set(y_pred)))
    cm = confusion_matrix(y_true, y_pred, labels=labels)
    cmn = cm.astype("float") / cm.sum(axis=1, keepdims=True)

    names = [id2label[i] for i in labels]
    fig, ax = plt.subplots(figsize=(5.5,5))
    im = ax.imshow(cmn, interpolation="nearest")
    ax.set_title("Confusion Matrix (normalized)")
    ax.set_xticks(range(len(labels))); ax.set_yticks(range(len(labels)))
    ax.set_xticklabels(names, rotation=25, ha="right"); ax.set_yticklabels(names)
    ax.set_xlabel("Predicted"); ax.set_ylabel("True")

    thr = cmn[~np.isnan(cmn)].max() / 2.0 if np.isfinite(cmn).any() else 0.5
    for i in range(cmn.shape[0]):
        for j in range(cmn.shape[1]):
            val = cmn[i, j]
            txt = f"{0.0:.2f}\n(0)" if not np.isfinite(val) else f"{val:.2f}\n({cm[i,j]})"
            ax.text(j, i, txt,
                    ha="center", va="center",
                    color="white" if (np.isfinite(val) and val > thr) else "black",
                    fontsize=9)
    fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
    plt.tight_layout(); plt.savefig(out_path, dpi=160); plt.show()

# ------------------ MAIN ------------------
def main():
    ensure_out()

    # Load splits and detect columns
    raw = {
        "train": read_csv(TRAIN_PATH),
        "valid": read_csv(VALID_PATH),
        "test":  read_csv(TEST_PATH),
    }

    texts_raw: Dict[str, pd.Series] = {}
    labels_raw: Dict[str, pd.Series] = {}
    used_cols: Dict[str, Dict[str,str]] = {}

    for split, df in raw.items():
        text, lab, tcol, lcol = detect_text_and_label(df)
        texts_raw[split] = text
        labels_raw[split] = lab.loc[text.index]
        used_cols[split] = {"text_col": tcol, "label_col": lcol}

    # Build consistent label map across all splits
    label2id = build_label_map(labels_raw)
    id2label = {int(v): str(k) for k, v in label2id.items()}

    # Apply mapping & finalize frames
    splits = {}
    for split in ["train", "valid", "test"]:
        t = texts_raw[split]
        y = apply_label_map(labels_raw[split], label2id)
        df = pd.DataFrame({"text": t, "label": y}).dropna().reset_index(drop=True)
        splits[split] = df

    # Prepare data
    X_train = pd.concat([splits["train"]["text"], splits["valid"]["text"]], axis=0).values
    y_train = pd.concat([splits["train"]["label"], splits["valid"]["label"]], axis=0).values
    X_test  = splits["test"]["text"].values
    y_test  = splits["test"]["label"].values

    # Build pipeline
    pipe = build_pipeline()

    # --- Learning curve (on train+valid with CV) ---
    lc_path = os.path.join(OUT_DIR, "twh_accuracy_curve.png")
    plot_learning_curve(pipe, X_train, y_train, lc_path)

    # --- Fit & evaluate ---
    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)

    # Metrics
    acc = accuracy_score(y_test, y_pred)
    p, r, f1, _ = precision_recall_fscore_support(y_test, y_pred, average="macro", zero_division=0)

    # Confusion matrix plot
    cm_path = os.path.join(OUT_DIR, "twh_confusion_matrix.png")
    plot_confusion_matrix(y_test, y_pred, id2label, cm_path)

    # Reports & artifacts
    report_txt = classification_report(
        y_test, y_pred,
        target_names=[id2label[i] for i in sorted(set(y_test) | set(y_pred))],
        zero_division=0
    )
    with open(os.path.join(OUT_DIR, "twh_classification_report.txt"), "w", encoding="utf-8") as f:
        f.write(report_txt)

    # Save metrics JSON
    metrics = {
        "created_utc": datetime.utcnow().isoformat() + "Z",
        "sources": {
            "train": TRAIN_PATH, "valid": VALID_PATH, "test": TEST_PATH
        },
        "text_cols_used": used_cols,
        "label2id": {str(k): int(v) for k, v in label2id.items()},
        "id2label": {str(k): v for k, v in id2label.items()},
        "sizes": {k: int(len(v)) for k, v in splits.items()},
        "test_metrics": {
            "accuracy": round(float(acc), 4),
            "macro_precision": round(float(p), 4),
            "macro_recall": round(float(r), 4),
            "macro_f1": round(float(f1), 4),
        },
        "artifacts": {
            "accuracy_curve_png": os.path.join(OUT_DIR, "twh_accuracy_curve.png"),
            "confusion_matrix_png": os.path.join(OUT_DIR, "twh_confusion_matrix.png"),
            "classification_report_txt": os.path.join(OUT_DIR, "twh_classification_report.txt"),
            "predictions_csv": os.path.join(OUT_DIR, "twh_predictions_test.csv"),
        }
    }
    with open(os.path.join(OUT_DIR, "twh_metrics.json"), "w", encoding="utf-8") as f:
        json.dump(metrics, f, ensure_ascii=False, indent=2)

    # Save per-row predictions (test)
    pred_df = pd.DataFrame({
        "text": X_test,
        "y_true": [id2label[int(i)] for i in y_test],
        "y_pred": [id2label[int(i)] for i in y_pred]
    })
    pred_df.to_csv(os.path.join(OUT_DIR, "twh_predictions_test.csv"), index=False, encoding="utf-8")

    # Console summary
    print("=== DONE ===")
    print("Acc:", round(acc,4), "| Macro-F1:", round(f1,4))
    print("Saved:",
          "twh_accuracy_curve.png, twh_confusion_matrix.png, twh_metrics.json,",
          "twh_classification_report.txt, twh_predictions_test.csv", sep="\n")

if __name__ == "__main__":
    main()
```
